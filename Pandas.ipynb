```python
df = pd.DataFrame({
    'sales': [300, 200, 300, 900, 1000, 600],
    'signups': [500, 500, 600, 1200, 1400, 1300],
    'visits': [2000, 4200, 2800, 6200, 8100, 5000],
}, index=pd.date_range(start='2018/01/01', end='2018/07/01',
                       freq='M'))
```

```python
df['visits'].apply(lambda x: f'{x:,}')
```

```python
# line by index value, return a DataFrame
df.loc[['2018-06-30']]
```
